```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df1 = df.replace({'M':0,'R':1})
```

```python
df1.head(10)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
X = df1.drop([60],axis = 'columns')
y = df1[60]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state= 1)
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'  ),
    keras.layers.Dense(30,activation='relu'  ),
    keras.layers.Dense(15,activation='relu'  ),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
    )

model.fit(X_train,y_train,epochs = 100,batch_size=8)
```

```python
model_a_accuracy = model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred[:10]
```

```python
y_pred = np.round(y_pred)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
classification_report(y_test,y_pred)
```

```python
model_b = keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'  ),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'  ),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'  ),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1,activation='sigmoid')
])

model_b.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
    )

model_b.fit(X_train,y_train,epochs = 100,batch_size=8)
```

```python
model_b_accuracy = model_b.evaluate(X_test,y_test)
```

```python
print("Model A accuracy" ,model_a_accuracy[1])
print(" ")
print("after dropput , Dropping some hidden layers for better training ")
print(" ")
print("Model B accuracy" ,model_b_accuracy[1])
```
